```python
import numpy as np
import random
import pandas as pd
```

```python
np.random.seed(42)
df1 = pd.DataFrame(np.random.rand(2048,20))
df1
```

```python
np.random.seed(38)
df2 = pd.DataFrame(10*np.random.rand(2048,20))
df2
```

```python
df1.to_csv("EEM1.csv")
df2.to_csv("EEM10.csv")
```

```python
x_train = [df1, df2]
y_train = [1, 10]
```

```python
x_train = np.array(x_train)
x_train.reshape(2, -1).shape
# np.array(df1).flatten().shape
```

```python
x_train.reshape(2,-1).shape
np.array(y_train).shape
```

```python
from sklearn.linear_model import LinearRegression
logreg = LinearRegression().fit(x_train.reshape(2,-1), np.array(y_train))
print("Training set score : {:.3f}".format(logreg.score(x_train.reshape(2,-1), y_train)))
```

```python
logreg.predict(np.array(df2).reshape(1, -1))
```

```python
logreg.coef_
```

```python
logreg.intercept_
```
